```python
import tpot
import sklearn
import sklearn.datasets
import sklearn.model_selection
import timeit
from tpot import TPOTClassifier
import pandas as pd
print('tpot: %s' % tpot.__version__)
```

```python
def dataSetup(name):
  X,y = sklearn.datasets.fetch_openml(name, as_frame=True, return_X_y=True)
  train = pd.concat([X, y], axis=1, join='inner')
  train.dropna()
  train = train.apply(lambda x: pd.factorize(x)[0])
  X,y = train.iloc[:,:-1], train.iloc[:, -1]
  # X = X.apply(lambda x: pd.factorize(x)[0])
  X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y,random_state=42)
  return (X_train, X_test, y_train, y_test)
```

```python
X_train, X_test, y_train, y_test = dataSetup("spambase")
X_train.head()
```

```python
model = TPOTClassifier(max_time_mins = 60,generations = 5,scoring='accuracy', verbosity=2, random_state=1, n_jobs=-1)
```

```python
start = timeit.default_timer()
model.fit(X_train, y_train)
stop = timeit.default_timer()
print(model.score(X_test,y_test))
model.export('tpot_sonar_best_model.py')
```
